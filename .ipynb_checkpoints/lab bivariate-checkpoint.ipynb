```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import chi2_contingency
from scipy.stats import cramers_v
```

```python
# loading the dataset
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
df.head()
```

```python
# Part 1: Analyzing Best-Seller Trends Across Product Categories
# 1 Crosstab between product category and isBestSeller
category_isbestseller_crosstab = pd.crosstab(df['category'], df['isBestSeller'])

# Calculate the proportion of best-sellers for each category
category_isbestseller_proportion = category_isbestseller_crosstab.div(category_isbestseller_crosstab.sum(axis=1), axis=0)

# Sort the categories based on the proportion of best-sellers in descending order
sorted_categories = category_isbestseller_proportion.sort_values(by=True, ascending=False)

print(sorted_categories)
```

```python
# 2 statistical test
# Crosstab between product category and isBestSeller
category_isbestseller_crosstab = pd.crosstab(df['category'], df['isBestSeller'])

# Chi-square test for independence
chi2_stat, p_val, _, _ = chi2_contingency(category_isbestseller_crosstab)
# Compute Cramér's V
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# Compute Cramér's V
cramers_v_score = cramers_v(category_isbestseller_crosstab.values)

print("Cramér's V:", cramers_v_score)
```

```python
# visualizations 
# Plotting the stacked bar chart with increased figure size
plt.figure(figsize=(16, 10))  
ax = category_isbestseller_crosstab.plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'])
plt.title('Relationship Between Product Categories and Best-Seller Status')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.legend(title='Best Seller', labels=['Not Best Seller', 'Best Seller'])
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Part 2: Exploring Product Prices and Ratings Across Categories and Brands
# 0 Preliminary Step: Remove outliers in product prices.
# Calculate Q1, Q3, and IQR for product prices
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the dataset
df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

# Display the shape of the original and outlier-removed datasets
print("Original dataset shape:", df.shape)
print("Dataset shape after removing outliers:", df_no_outliers.shape)
```

```python
# 1 Violin Plots
# Filter out the top 20 categories based on count
top_categories = df['category'].value_counts().nlargest(20).index
df_top_categories = df[df['category'].isin(top_categories)]

# Create a violin plot for the filtered categories
plt.figure(figsize=(12, 8))
sns.violinplot(x='category', y='price', data=df_top_categories)
plt.title('Distribution of Price Across Top 20 Product Categories')
plt.xlabel('Category')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Determine the product category with the highest median price
category_median_prices = df.groupby('category')['price'].median()
category_with_highest_median_price = category_median_prices.idxmax()
highest_median_price = category_median_prices.max()

print("Product category with the highest median price:", category_with_highest_median_price)
print("Highest median price:", highest_median_price)
```

```python
# 2 Bar Charts
# Calculate the average price of products for each of the top 10 categories
category_avg_prices = df_top_categories.groupby('category')['price'].mean()

# Create a bar chart to compare the average prices
plt.figure(figsize=(10, 6))
category_avg_prices.sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Price of Products for Top 10 Product Categories')
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Determine the product category with the highest average price
category_with_highest_avg_price = category_avg_prices.idxmax()
highest_avg_price = category_avg_prices.max()

print("Product category with the highest average price:", category_with_highest_avg_price)
print("Highest average price:", highest_avg_price)
```

```python
# Box Plots
# Filter out the top 10 categories based on count
top_categories = df['category'].value_counts().nlargest(10).index
df_top_categories = df[df['category'].isin(top_categories)]

# Create side-by-side box plots for the ratings of products in these categories
plt.figure(figsize=(12, 8))
sns.boxplot(x='category', y='stars', data=df_top_categories)
plt.title('Distribution of Product Ratings Across Top 10 Product Categories')
plt.xlabel('Category')
plt.ylabel('Stars')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Determine the category with the highest median rating
category_median_ratings = df.groupby('category')['stars'].median()
category_with_highest_median_rating = category_median_ratings.idxmax()
highest_median_rating = category_median_ratings.max()

print("Category with the highest median rating:", category_with_highest_median_rating)
print("Highest median rating:", highest_median_rating)
```

```python
# Part 3: Investigating the Interplay Between Product Prices and Ratings
# Calculate the correlation coefficient between price and stars
correlation_coefficient = df['price'].corr(df['stars'])

print("Correlation coefficient between price and stars:", correlation_coefficient)
# Typically, a correlation coefficient closer to 1 or -1 indicates a stronger correlation.
```

```python
# 2 visualizations
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['stars'], df['price'], alpha=0.5)
plt.title('Relationship between Product Rating and Price')
plt.xlabel('Stars')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
 # Calculate correlation matrix
# Exclude non-numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
corr_matrix = numeric_columns.corr()

# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# Create QQ plot for product prices
plt.figure(figsize=(8, 6))
stats.probplot(df['price'], dist="norm", plot=plt)
plt.title('QQ Plot for Product Prices')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid(True)
plt.show()
```
